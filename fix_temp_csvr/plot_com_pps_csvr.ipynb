```python
import os 
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_formats=['svg']
%reset_selective -f a

START = 100001
STOP = 200002
STEP = 100 # * 10fs
##
LIST_LEGEND = [
    'DOF 3N-3 no framwork',
    'DOF 3N-3 no framwork zero momentum',
    'DOF 3N with framework',
]
SIMULATION = [
    'noframe_dof3',
    'noframe_dof3_fixP',
    'frame_dof0',
]
fig, axs=plt.subplots(2,1,figsize=[8.4,6.4],sharex=True)
for i, i_run in enumerate(SIMULATION):
    data=np.loadtxt('%s/thermo.txt'%i_run,dtype=float)
    axs[0].plot(data[START:STOP:STEP,1],data[START:STOP:STEP,5])
    axs[1].plot(data[START:STOP:STEP,1],data[START:STOP:STEP,6]-i*0.01)
axs[1].set_xlabel('$\Delta t$ (10 fs)')
axs[0].set_ylabel('position$_\mathrm{COM}$ ($\AA$)')
axs[1].set_ylabel('velocity$_\mathrm{COM}$ ($\AA^2$/fs)')
axs[0].set_ylim([-0,300])
axs[0].legend(LIST_LEGEND,loc='lower left')
plt.show()
plt.close()
```
